```python
dir = '/content/aalto_summer24_project/data/filtered_data.csv'
df = pd.read_csv(dir)
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe().round(2).T
```

```python
X_train_columns = [ 'MED', 'PIMP', 'TP60', #'NBMENFISC', 'NBPERSMENFISC',
       'TP60AGE1', 'TP60AGE2', 'TP60AGE3', 'TP60AGE4', 'TP60AGE5', 'TP60AGE6',
       'TP60TOL1', 'TP60TOL2', 'PACT', 'PTSA', 'PCHO', 'PBEN', 'PPEN', 'PPAT',
       'PPSOC', 'PPFAM', 'PPMINI', 'PPLOGT', 'PIMPOT', 'D1', 'D9', 'RD' ] #,'lat','lng'


X = df[X_train_columns]

X.head(2)
```

```python
X = X.fillna(X.median())
```

```python
from sklearn.cluster import KMeans

# Initialize KMeans with 5 clusters
kmeans = KMeans(n_clusters=5, random_state=42)

# Fit the model to your data
kmeans.fit(X)

# Get the cluster labels for each data point
labels = kmeans.labels_

# Add the labels to your DataFrame
df['cluster'] = labels
```

```python
import plotly.express as px

# Create the map, setting the scope to France
fig = px.scatter_geo(
    df,
    lat='lat',
    lon='lng',
    size='TP60',
    color='cluster',  # Color points by cluster
    hover_name='LIBGEO',
    projection="natural earth",
    scope='europe'
)
fig.update_geos(
    lataxis_range=[41, 51],
    lonaxis_range=[-5, 9]
)
fig.update_layout(
    title='Clusters on Map'
)
fig.show()
```

```python
from sklearn.metrics import silhouette_score

# Calculate Silhouette Score
silhouette_avg = silhouette_score(X, labels)
print("Silhouette Score:", silhouette_avg)

# Example: Trying different number of clusters
for n_clusters in range(2, 11):
  kmeans = KMeans(n_clusters=n_clusters, random_state=42)
  kmeans.fit(X)
  labels = kmeans.labels_
  silhouette_avg = silhouette_score(X, labels)
  print(f"For n_clusters = {n_clusters}, the average silhouette_score is : {silhouette_avg}")

# Example: Trying feature selection (simple example, needs refinement)
from sklearn.feature_selection import SelectKBest, f_classif
selector = SelectKBest(f_classif, k=10)  # Select top 10 features
X_new = selector.fit_transform(X, labels)

kmeans = KMeans(n_clusters=5, random_state=42)
kmeans.fit(X_new)
labels_new = kmeans.labels_
silhouette_avg_new = silhouette_score(X_new, labels_new)
print("Silhouette Score after feature selection:", silhouette_avg_new)
```

```python
df['cluster_new'] = labels_new

# Create the map for the new clusters
fig_new = px.scatter_geo(
    df,
    lat='lat',
    lon='lng',
    size='TP60',
    color='cluster_new',  # Color points by the new cluster labels
    hover_name='LIBGEO',
    projection="natural earth",
    scope='europe'
)
fig_new.update_geos(
    lataxis_range=[41, 51],
    lonaxis_range=[-5, 9]
)
fig_new.update_layout(
    title='Clusters on Map (After Feature Selection)'
)
fig_new.show()
```

```python
# prompt: group by clusters and do not include non numerical columns

# Select numerical columns and the 'cluster' column
numerical_cols = df.select_dtypes(include=['number']).columns
df_numerical = df[numerical_cols.union(['cluster_new'])]

# Group by clusters and calculate mean for each numerical column
cluster_means = df_numerical.groupby('cluster_new').mean()

cluster_means.T.round(2)
```
