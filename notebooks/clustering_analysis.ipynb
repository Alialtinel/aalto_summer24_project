```python
dir = '/content/aalto_summer24_project/data/filtered_data.csv'
df = pd.read_csv(dir)
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.dropna(subset=['lat','lng'], inplace=True)
```

```python
df.describe().round(2).T
```

```python
X_train_columns = [ 'MED', 'PIMP', 'TP60', #'NBMENFISC', 'NBPERSMENFISC',
       'TP60AGE1', 'TP60AGE2', 'TP60AGE3', 'TP60AGE4', 'TP60AGE5', 'TP60AGE6',
       'TP60TOL1', 'TP60TOL2', 'PACT', 'PTSA', 'PCHO', 'PBEN', 'PPEN', 'PPAT',
       'PPSOC', 'PPFAM', 'PPMINI', 'PPLOGT', 'PIMPOT', 'D1', 'D9', 'RD' ] #,'lat','lng'


X = df[X_train_columns]

X.head(2)
```

```python
X = X.fillna(X.median())
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans

# Initialize KMeans with 5 clusters
kmeans = KMeans(n_clusters=5, random_state=42)

# Fit the model to your data
kmeans.fit(X_scaled)

# Get the cluster labels for each data point
labels = kmeans.labels_

# Add the labels to your DataFrame
df['cluster'] = labels
```

```python
import plotly.express as px

# Create the map, setting the scope to France
fig = px.scatter_geo(
    df,
    lat='lat',
    lon='lng',
    size='TP60',
    color='cluster',  # Color points by cluster
    hover_name='LIBGEO',
    projection="natural earth",
    scope='europe'
)
fig.update_geos(
    lataxis_range=[41, 51],
    lonaxis_range=[-5, 9]
)
fig.update_layout(
    title='Clusters on Map'
)
fig.show()
```

```python
# Select numerical columns and the 'cluster' column
numerical_cols = df.select_dtypes(include=['number']).columns
df_numerical = df[numerical_cols.union(['cluster'])]

# Group by clusters and calculate mean for each numerical column
cluster_means = df_numerical.groupby('cluster').mean()

cluster_means.T.round(2)
```
