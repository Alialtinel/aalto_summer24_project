```python
df_2015 = pd.read_excel('/content/aalto_summer24_project/data/french_poverty_urban_areas_2015.xlsx')
df_2016 = pd.read_excel('/content/aalto_summer24_project/data/french_poverty_urban_areas_2016.xlsx')
df_2017 = pd.read_excel('/content/aalto_summer24_project/data/french_poverty_urban_areas_2017.xlsx')
```

```python

import plotly.express as px

# 2015
corr_matrix_2015 = df_2015.select_dtypes(include='number').corr()
fig_2015 = px.imshow(corr_matrix_2015,
                    color_continuous_scale='RdBu',
                    title='Correlation Heatmap 2015')
fig_2015.show()

# 2016
corr_matrix_2016 = df_2016.select_dtypes(include='number').corr()
fig_2016 = px.imshow(corr_matrix_2016,
                    color_continuous_scale='RdBu',
                    title='Correlation Heatmap 2016')
fig_2016.show()

# 2017
corr_matrix_2017 = df_2017.select_dtypes(include='number').corr()
fig_2017 = px.imshow(corr_matrix_2017,
                    color_continuous_scale='RdBu',
                    title='Correlation Heatmap 2017')
fig_2017.show()
```
